# Conversoes Numericas com Binario

Notebook em 4 etapas para estudar niveis de quantizacao e conversoes entre bases numericas usando Python.

Estrutura do trabalho:
- Etapa 1: niveis de quantizacao, bits e combinacoes
- Etapa 2: decimal inteiro para binario
- Etapa 3: decimal fracionario para binario e vice-versa
- Etapa 4: octal e hexadecimal para binario e vice-versa, com conversao para decimal usando o binario como intermediario


```python
from conversoes_binarias import (
    niveis_para_bits_e_combinacoes,
    decimal_inteiro_para_binario,
    decimal_fracionario_para_binario,
    binario_fracionario_para_decimal,
    octal_para_binario,
    binario_para_octal,
    hexa_para_binario,
    binario_para_hexa,
    octal_para_decimal_via_binario,
    hexa_para_decimal_via_binario,
)

print('Modulo carregado com sucesso.')
```

## Etapa 1 - Niveis de quantizacao

Nesta etapa, o usuario informa o numero de niveis de quantizacao. O programa calcula o numero minimo de bits necessarios e mostra todas as combinacoes possiveis desses bits.


```python
niveis = int(input('Digite o numero de niveis de quantizacao: '))
bits, combinacoes = niveis_para_bits_e_combinacoes(niveis)

print(f'Numero minimo de bits: {bits}')
print('Combinacoes possiveis:')
for combinacao in combinacoes:
    print(combinacao)
```

## Etapa 2 - Decimal inteiro para binario

Aqui e feita a conversao de um numero inteiro decimal para binario, sem usar a funcao `bin()`.


```python
numero_decimal = int(input('Digite um numero inteiro decimal: '))
numero_binario = decimal_inteiro_para_binario(numero_decimal)
print(f'{numero_decimal} em binario = {numero_binario}')
```

## Etapa 3 - Decimal fracionario para binario e vice-versa

Nesta parte, o notebook converte um numero decimal fracionario para binario e tambem recebe um numero binario fracionario para converter de volta para decimal.


```python
decimal_fracionario = input('Digite um numero decimal fracionario (ex.: 10.625): ')
precisao = int(input('Digite a quantidade maxima de bits da parte fracionaria: '))
binario_gerado = decimal_fracionario_para_binario(decimal_fracionario, precisao=precisao)
print(f'{decimal_fracionario} em binario = {binario_gerado}')

binario_fracionario = input('Digite um numero binario fracionario (ex.: 1010.101): ')
decimal_convertido = binario_fracionario_para_decimal(binario_fracionario)
print(f'{binario_fracionario} em decimal = {decimal_convertido}')
```

## Etapa 4 - Octal e hexadecimal para binario e vice-versa

O objetivo desta etapa e converter valores octais e hexadecimais para binario, fazer o caminho inverso, e ainda obter o valor decimal usando o binario como intermediario.


```python
valor_octal = input('Digite um numero octal: ')
binario_octal = octal_para_binario(valor_octal)
decimal_octal = octal_para_decimal_via_binario(valor_octal)
print(f'Octal {valor_octal} -> Binario {binario_octal}')
print(f'Binario {binario_octal} -> Octal {binario_para_octal(binario_octal)}')
print(f'Octal {valor_octal} -> Decimal {decimal_octal} (via binario)')

valor_hexa = input('Digite um numero hexadecimal: ')
binario_hexa = hexa_para_binario(valor_hexa)
decimal_hexa = hexa_para_decimal_via_binario(valor_hexa)
print(f'Hexa {valor_hexa} -> Binario {binario_hexa}')
print(f'Binario {binario_hexa} -> Hexa {binario_para_hexa(binario_hexa)}')
print(f'Hexa {valor_hexa} -> Decimal {decimal_hexa} (via binario)')
```
